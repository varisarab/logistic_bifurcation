```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logistic(r, x):
    return r * x * (1 - x)
```

```python
n = 1000000  # total number of points
r = np.linspace(2.5, 4.0, n)
iterations = 1000
```

```python
x = np.linspace(0, 1)
fig, ax = plt.subplots(1, 1)
ax.plot(x, logistic(2, x), 'k')
```

```python
def system(r, x0, n, ax=None):
    t = np.linspace(0, 1)
    ax.plot(t, logistic(r, t), 'k', lw=2)
    ax.plot([0, 1], [0, 1], 'k', lw=2)

    x = x0
    for i in range(n):
        y = logistic(r, x)
        #plot 2 lines
        ax.plot([x, x], [x, y], 'k', lw=1)
        ax.plot([x, y], [y, y], 'k', lw=1)
        #increasing opacity
        ax.plot([x], [y], 'ok', ms=10, alpha=(i + 1) / n)
        x = y
        
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_title(f"$r={r:.1f}, \, x_0{x0:.1f}$")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6), sharey=True)
system(2.5, .1, 10, ax=ax1)
system(3.5, .1, 10, ax=ax2)
```

```python
x = np.random.uniform(0.0, 1.0, n)  # initial value
for _ in range(iterations):  # iterate
    x = logistic(r, x)

fig = plt.figure(figsize=(20, 10))
ax = plt.axes()

ax.set_title("Bifurcation Diagram", fontsize=24)
ax.set_xlabel('r', fontsize=24)
ax.set_ylabel('$x^{(r)}_{n}$', fontsize=24)
ax.tick_params(axis='both', which='major', labelsize=18)
ax.set_xlim(2.5, 4.0)
ax.set_ylim(0, 1.0)

ax.plot(r, x, ',b', alpha=.5, color="green")
plt.savefig('bifurcation_3.png', bbox_inches='tight')
```
